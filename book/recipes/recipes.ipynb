المكان والزمان:

- الوقت المتبقي للصلاة
- عدد الأيام بين تاريخين (وكذلك عدد الساعات والدقائق والثواني)
- حساب المسافة بين نقطتين جغرافيتين

اقتصاد:

- توزيع موارد على مصارف (activity selection algorithm)
- رسم شجرة اعتمادية مواد على بعضها (مثل اعتماد مادة دراسية على مجموعة من المواد)

ربط مع نظام التشغيل:

- توقيت برنامج ليعمل في الساعة التاسعة كل يوم: مثلاً: تحديث النسخة الاحتياطية للجهاز
- ربط البرنامج ليعمل بمفتاح مختصر (shortcut)
- عمل برنامج يشغل لك برامج أخرى .. مثلاً لتجهيز مساحة العمل

ربط مع أنظمة أخرى (API) IFTTT:

- إرسال إيميل
- عرض تغريدات من مساحات معينة
