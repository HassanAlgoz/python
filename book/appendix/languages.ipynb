
ومن اللغات ما هو واسع جدًّا من ناحية الاستعمال إلا أن استعمالها محدد لمجال معيَّن وليسَت لغة عامة يُكتب بها أي شيء؛ وهي **اللغات الوصفية** (Declarative Languages) وللتكميل نضيفها في الرسم:

```{mermaid}

            Query2[SQL]
        end

        end

        Script1[Bash]
```
