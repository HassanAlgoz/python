### 9.1 Exercise: Anagram test
Main idea: Sorting both words’ letters makes anagram comparison straightforward.

Task: Write `is_anagram(word1, word2)` that returns `True` if the two strings are anagrams. Complete the doctest outline in the next cell, then search the word list for anagrams of `takes` as shown below.

**Setup:** The next code cell(s) contain starter outlines and/or doctests to complete.


To get you started, here's an outline of the function with doctests.

```python
def is_anagram(word1: str, word2: str) -> bool:
    """Checks whether two words are anagrams.

    >>> is_anagram('tops', 'stop')
    True
    >>> is_anagram('skate', 'takes')
    True
    >>> is_anagram('tops', 'takes')
    False
    >>> is_anagram('skate', 'stop')
    False
    """
    word1 = word1.lower()
    word2 = word2.lower()
    return sorted(word1) == sorted(word2)
```

You can use `doctest` to test your function.

```python
from collections.abc import Callable
from doctest import run_docstring_examples

def run_doctests(func: Callable[..., object]) -> None:
    run_docstring_examples(func, globals(), name=func.__name__)
```

```python
run_doctests(is_anagram)
```

Using your function and the word list, find all the anagrams of `takes`.

```python
target = "takes"
anagrams = set() # this will hold the anagrams we find

# Open the file and read it line by line
with open('words.txt', 'r') as f:
    for line in f:
        # need to strip it so i don't have to worry about the newline character
        word = line.strip()
        
        # Check if it's an anagram
        if is_anagram(word, target):
            anagrams.add(word)

anagrams = anagrams - {target}
print(anagrams)
```

So we can write a function that reverses a word like this.

```python
def reverse_word(word: str) -> str:
    return ''.join(reversed(word))
```

A palindrome is a word that is spelled the same backward and forward, like "noon" and "rotator".
Write a function called `is_palindrome` that takes a string argument and returns `True` if it is a palindrome and `False` otherwise.

Here's an outline of the function with doctests you can use to check your function.

```python
def is_palindrome(word: str) -> bool:
    """Check if a word is a palindrome.

    >>> is_palindrome('bob')
    True
    >>> is_palindrome('alice')
    False
    >>> is_palindrome('a')
    True
    >>> is_palindrome('')
    True
    """
    return word == reverse_word(word)
```

```python
run_doctests(is_palindrome)
```

**Example:** Use the following loop to find all of the palindromes in the word list with at least 7 letters.

```python
with open('words.txt') as f:
    word_list = f.readlines()

for word in word_list:
    if len(word) >= 7 and is_palindrome(word):
        print(word)
```

### 9.3 Exercise: Reverse word order in a sentence
Main idea: Split into words, reverse the list, then join; adjust capitalization as needed.

Task: Implement `reverse_sentence(s)` so words appear in reverse order with the formatting rules in the doctests (see hint about `capitalize` in the chapter).

**Setup:** The next code cell(s) contain starter outlines and/or doctests to complete.


To get you started, here's an outline of the function with doctests.

```python
def reverse_sentence(input_string: str) -> str:
    '''Reverse the words in a string and capitalize the first.

    >>> reverse_sentence('Reverse this sentence')
    'Sentence this reverse'

    >>> reverse_sentence('Python')
    'Python'

    >>> reverse_sentence('')
    ''

    >>> reverse_sentence('One for all and all for one')
    'One for all and all for one'
    '''
    words = input_string.split()
    words.reverse()
    return ' '.join(words).capitalize()
```

```python
run_doctests(reverse_sentence)
```

### 9.4 Exercise: Total length of a word list
Main idea: Accumulate string lengths from a file or list.

Task: Write `total_length` (or follow the chapter’s naming) so the sum of lengths of all words in `word_list` matches 902,728 as stated.


```python
def total_length(file: str) -> int:
    total = 0
    with open(file, 'r') as f:
        for line in f:
            word = line.strip()
            total += len(word)
    return total

result = total_length('words.txt')
print(result)
```
