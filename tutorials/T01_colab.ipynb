# Introduction to Notebooks


- Double click a cell/block to edit it
- `Ctrl + Enter` to execute it
- `Shift + Enter` to execute it and jump to the next one
- Change cell type from the top toolbar (code, markdown, raw)
    - `code` is executable Python code
    - `markdown` is like this cell (has special rendering of text)
    - `raw` unlike markdown, displays the text exactly as it is written, without any modification or rendering.


- Escape mode:
    - `a`: create a cell above selected cell
    - `b`: create a cell below selected cell
    - `dd`: delete selected cell(s)
    - `x`: cut selected cell(s)
    - `v`: paste cell(s)
    - `m`: convert cell to markdown
    - `y`: convert cell to code

## Markdown

**Markdown** is a lightweight *markup* language that allows you to create formatted text using a ~plain~ text editor. It is commonly used to create:

1. README files
2. documentation
3. blog posts

A good reference for markdown syntax is: https://www.markdownguide.org/basic-syntax/

```md
# This is a level 1 heading
## This is a level 2 heading
### This is a level 3 heading

**This is bold text.**

*This is italic text.*

- This is an item in an unordered list.
* This is another item in an unordered list.

1. This is an item in an ordered list.
1. This is another item in an ordered list.

Link to [Goo gell](https://www.google.com)

A table:

| Name | Age |
|---|---|
| Alice | 25 |
| Bob | 30 |
| Carol | 35 |
```

### Equations (LaTeX)

Notebooks' Markdown also supports *LaTeX* formatted equations, in two ways: inline (with `$`) and block (with `$$`).

Example of inline equation: $E = mc^2$

Example of block equation:

$$
f(x) = \frac{y^2+2}{z}
$$

# Plots

````python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data:
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot the data:
plt.plot(x, y)

# Set the title and labels:
plt.title("Sine Wave")
plt.xlabel("X")
plt.ylabel("Y")

# Show the plot:
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

labels = ["A", "B", "C"]
values = [10, 20, 30]

plt.bar(labels, values)
plt.xlabel("Labels")
plt.ylabel("Values")
plt.title("Bar Plot")
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(1000)

plt.hist(x)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(1000)
y = np.random.rand(1000)

plt.scatter(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot")
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

labels = ["A", "B", "C"]
values = [10, 20, 30]

plt.pie(values, labels=labels, autopct="%1.1f%%")
plt.title("Pie Chart")
plt.show()
````
